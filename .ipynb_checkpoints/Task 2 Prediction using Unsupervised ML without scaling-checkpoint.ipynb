```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from math import pi 
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder 
from sklearn.cluster import KMeans

%matplotlib inline
```

```python
df = pd.read_csv('./Iris.csv')
```

```python
df.head()
```

# data preprocessing

```python
df.drop('Id',axis=1,inplace=True)
```

```python
le = LabelEncoder()
df['Species'] = le.fit_transform(df['Species'])
```

# feature engineering 

```python
#in this part we can add 2 features on accordance to the sepal volume (volume of a cylinder) 
# and the the surface area of a cut on the petal 
```

```python
#cylinder volume = r^2 * pi * hight
df['sepal_vol_cm3'] = (( (df['SepalWidthCm']/2) **2)*pi)  *  df['SepalLengthCm'] 

df['petal_sidesurface_air']= df['PetalLengthCm'] * df['PetalWidthCm']
```

```python
scores = []
range_values = range(1,10)

for i in range_values:
    kmeans = KMeans(n_clusters = i)
    kmeans.fit(df) 
    scores.append(kmeans.inertia_)

plt.plot(scores,'bx-')
```

# without PCA 

```python
kmeans = KMeans(3,random_state=42)
kmeans.fit(df)
y_pred = kmeans.predict(df)
```

```python
print(f'inertia = {kmeans.inertia_}')
```

```python
sns.scatterplot(x='sepal_vol_cm3',y='petal_sidesurface_air',data=df ,
                hue=y_pred,palette=['red','green','blue'])
```

```python
df_final = df 
df_final['predicted'] = y_pred
sns.pairplot(df_final,hue='predicted',palette=['r','g','b'])
```

```python
sns.histplot(data=df_final,x='predicted',hue='Species',palette=['r','g','b'])
```

# after PCA 

```python
# PCA has the role of englobing the original data using less columns 
# thus less variables with almost the same amount of data 
```

```python
pca = PCA(n_components=3,random_state=42)
pca.fit(df)
```

```python
print(f'{sum(pca.explained_variance_ratio_):.4f} of the data is kept')
```

```python
#99 of the informations from df is now located in only 3 columns 
```

```python
df_pca = pd.DataFrame(pca.transform(df),columns=['pca1','pca2','pca3'])
```

```python
df_pca
```

```python
kmeans_pca = KMeans(n_clusters=3,random_state=42)
kmeans_pca.fit(df_pca)
y_pred = kmeans_pca.predict(df_pca)
df_pca['predict'] = y_pred
```

```python
print(f'inertia = {kmeans_pca.inertia_}')
```

```python
sns.pairplot(df_pca,hue='predict',palette=['r','g','b'])
```

```python
print(f'inertia without PCA= {kmeans.inertia_  :.4f}')
print(f'inertia with PCA = {kmeans_pca.inertia_ :.4f}')
```

```python
#using PCA helped us decrease the inirtia
```
